<b> Projek Data Customer </b> <br>
Buatlah contoh data customer sebanyak 1000 yang akan digunakan untuk prediksi:<br>
<ol>
    <li> Usia : 13 – 35 </li>
    <li> Gender : L / P </li>
    <li> Kota Asal : Jakarta, Bandung, Surabaya, Medan </li>
    <li> Jenis Tiket : VIP, CAT1, CAT2, Festival </li>
    <li> Hari Konser : Hari 1, Hari 2, Hari 3 </li>
    <li> Kursi : Berdiri, Duduk </li>
    <li> Member Favorit	: Taeyong, Mark, Jaehyun, Haechan </li>
    <li> Merchandise : Ya, Tidak </li>
    <li> Rating Konser : 1–5 </li>
    <li> Kategori Penonton : NCTzen Baru, NCTzen Lama </li>
</ol>

```python
import numpy as np
import pandas as pd

n_data = 1000

usia = np.random.randint(13, 36, size=n_data)
gender = np.random.choice(['L','P'], size=n_data)
kotaAsal = np.random.choice(['Jakarta','Bandung','Surabaya','Medan'], size=n_data)
jenisTiket = np.random.choice(['VIP','CAT1','CAT2','Festival'], size=n_data)
hariKonser = np.random.choice(['Hari 1','Hari 2','Hari 3'], size=n_data)
kursi = np.random.choice(['Duduk','Berdiri'], size=n_data)
memberFavorit = np.random.choice(['Taeyong','Mark','Jaehyun','Haechan'], size=n_data)
merchandise = np.random.choice(['Ya','Tidak'], size=n_data)
rating = np.random.randint(1, 6, size=n_data)
kategori = np.random.choice(['NCTzen Baru','NCTzen Lama'], size=n_data)

df = pd.DataFrame({
    'Usia' : usia,
    'Gender' : gender,
    'Kota Asal' : kotaAsal,
    'Jenis Tiket' : jenisTiket,
    'Hari Konser' : hariKonser,
    'Kursi' : kursi,
    'Member Favorit' : memberFavorit,
    'Merchandise' : merchandise,
    'Rating Konser' : rating,
    'Kategori Penonton' : kategori
})

print(df.info)
print(f"Usia penonton termuda: {df.Usia.min()}")
print(f"Usia penonton tertua : {df.Usia.max()}")

jmlL = df[df['Gender']=='L']
jmlP = df[df['Gender']=='P']
print(f"Jumlah penonton laki-laki  : {jmlL.size/10}")
print(f"Jumlah penonton perempuan : {jmlP.size/10}")

jkt = df[df['Kota Asal']=='Jakarta']
bdg = df[df['Kota Asal']=='Bandung']
sby = df[df['Kota Asal']=='Surabaya']
mdn = df[df['Kota Asal']=='Medan']
print(f"Jumlah penonton asal Jakarta : {jkt.size/10}")
print(f"Jumlah penonton asal Bandung : {bdg.size/10}")
print(f"Jumlah penonton asal Surabaya: {sby.size/10}")
print(f"Jumlah penonton asal Medan   : {mdn.size/10}")

vip = df[df['Jenis Tiket']=='VIP']
cat1 = df[df['Jenis Tiket']=='CAT1']
cat2 = df[df['Jenis Tiket']=='CAT2']
fest = df[df['Jenis Tiket']=='Festival']
print(f"Jumlah penonton tiket VIP     : {vip.size/10}")
print(f"Jumlah penonton tiket CAT1    : {cat1.size/10}")
print(f"Jumlah penonton tiket CAT2    : {cat2.size/10}")
print(f"Jumlah penonton tiket Festival: {fest.size/10}")

hari1 = df[df['Hari Konser']=='Hari 1']
hari2 = df[df['Hari Konser']=='Hari 2']
hari3 = df[df['Hari Konser']=='Hari 3']
print(f"Jumlah penonton di Hari 1: {hari1.size/10}")
print(f"Jumlah penonton di Hari 2: {hari2.size/10}")
print(f"Jumlah penonton di Hari 3: {hari3.size/10}")

duduk = df[df['Kursi']=='Duduk']
berdiri = df[df['Kursi']=='Berdiri']
print(f"Jumlah penonton duduk  : {duduk.size/10}")
print(f"Jumlah penonton berdiri: {berdiri.size/10}")

taeyong = df[df['Member Favorit']=='Taeyong']
mark = df[df['Member Favorit']=='Mark']
jaehyun = df[df['Member Favorit']=='Jaehyun']
haechan = df[df['Member Favorit']=='Haechan']
print(f"Jumlah fans Taeyong : {taeyong.size/10}")
print(f"Jumlah fans Mark    : {mark.size/10}")
print(f"Jumlah fans Jaehyun : {jaehyun.size/10}")
print(f"Jumlah fans Haechan : {haechan.size/10}")

beli = df[df['Merchandise']=='Ya']
print(f"Jumlah penonton beli merchandise: {beli.size/10}")

print(f"Rata-rata rating konser: {df['Rating Konser'].mean():.2f}")

baru = df[df['Kategori Penonton']=='NCTzen Baru']
lama = df[df['Kategori Penonton']=='NCTzen Lama']
print(f"Jumlah NCTzen Baru: {baru.size/10}")
print(f"Jumlah NCTzen Lama: {lama.size/10}")

df.to_csv('dataKonserNCT.csv')
```
